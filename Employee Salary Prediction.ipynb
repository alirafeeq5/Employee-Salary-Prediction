# **Splitting** **the** **data**

```python

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(train_data, Y, test_size = 0.2, random_state = 0)

#X_train= train_data
test=test_data
#Y_train=Y
```

# **preprocessing**

```python
from sklearn import preprocessing

categorical = ['work-class', 'education', 'marital-status', 'position', 'relationship','race' ,'sex','native-country']
for feature in categorical:
        le = preprocessing.LabelEncoder()
        X_train[feature] = le.fit_transform(X_train[feature])
        X_test[feature] =  le.fit_transform(X_test[feature])
        test[feature] =  le.fit_transform(test[feature])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns)
X_test = pd.DataFrame(scaler.fit_transform(X_test), columns = X_test.columns)

test = pd.DataFrame(scaler.fit_transform(test), columns = test_data.columns)

# ca=['capital-gain','capital-loss']
# cb=scaler.fit_transform(X_train[ca])
# X_train[ca] = cb
# ct=scaler.fit_transform(X_test[ca])
# X_test[ca]=ct
```

# **Random** **Forest**

```python
random_forest = RandomForestClassifier(n_estimators=250,max_features=5)
random_forest.fit(X_train, Y_train)
Y_prediction = random_forest.predict(X_test)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
```

```python
print("Train ACC : ",acc_random_forest)
score = (accuracy_score(Y_test,Y_prediction)*100)
print("Test ACC : ",score)
c2=confusion_matrix(Y_test,Y_prediction)
print(c2)
```
